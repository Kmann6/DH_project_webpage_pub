```python
import pandas as pd

# import dataset
df = pd.read_csv('/home/kmann/DH_project_webpage_pub/historical-popculture-artifacts-american-history - Sheet1.csv')
print("all rows:")
df
```

```python
# Clean the dataset by making all values packaging_text to lowercase
df['packaging_text'] = df['packaging_text'].str.lower()
print("cleaned dataset:")
df
```

```python
# Make values in first_year_sold integers
df['first_year_sold'] = df['first_year_sold'].astype(int)
print("dataset with first_year_sold as integers:")
df
```

```python
# Change column name from 'first_year_sold' to 'year'
df.rename(columns={'first_year_sold': 'year'}, inplace=True)
print("dataset with column name changed to 'year':")
df
```

```python
# Remove puncuation from values in packaging_text
df['packaging_text'] = df['packaging_text'].str.replace('[^\w\s]', '', regex=True)
print("dataset with punctuation removed from packaging_text:")
df
```

```python
# Save dataset to new csv file
df.to_csv('/home/kmann/DH_project_webpage_pub/cleaned_historical_popculture_artifacts.csv', index=False)
print("Dataset saved to cleaned_historical_popculture_artifacts.csv")
```
